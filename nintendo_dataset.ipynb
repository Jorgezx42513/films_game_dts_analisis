```python
#valor minimo
print("El valor minimo de metacritic es:",nintendo["Metacritic"].min())

#valor maximo
print("El valor maximo de metacritic es:",nintendo["Metacritic"].max())

#Media 
print("El valor de la media es:",nintendo["Metacritic"].mean())

#mediana 
print("El valor de la mediana es:",nintendo["Metacritic"].median())
```

```python
sns.set_theme('notebook', palette = 'bright')
```

```python
fig, axes = plt.subplots(1,2, figsize=(11,5))

axes[0].hist(nintendo['Metacritic'])
axes[0].xaxis.set_major_formatter(FormatStrFormatter('%.1f'))
axes[0].set_title('Puntuacion Metacritic general')
axes[0].set_xlabel('puntuacion')
axes[0].set_ylabel('frecuencia')

axes[1].hist(nintendo['Rotten Tomatoes'])
axes[1].xaxis.set_major_formatter(FormatStrFormatter('%.1f'))
axes[1].set_title('Puntuacion Rotten tomatoes general')
axes[1].set_xlabel('puntuacion')
axes[1].set_ylabel('frecuencia')
```

```python
publishers  = nintendo.groupby(nintendo["Original game publisher"]).agg({'Worldwide box office': 'sum' ,'Metacritic' : 'mean', 'Rotten Tomatoes': 'mean' })
publishers
```

```python
fig, axes = plt.subplots()

ordered_val = publishers.sort_values(by='Metacritic')
range_ = range(1,len(publishers.index)+1)

axes.stem(ordered_val['Metacritic'])
axes.set_xticks(range_, )
```
